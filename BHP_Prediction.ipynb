```python
df2=df[['location','size','total_sqft','bath','balcony','price']]
```

```python
df2['BHK']=df2['size'].apply(lambda x: int(x.split(" ")[0]))
```

```python
df2=df2.drop(['size'],axis=1)
```

```python
df2
```

```python
df2['total_sqft'].unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
    
```

```python
df2[~df2['total_sqft'].apply(is_float)].head(20)
```

```python
def convert(x):
    tokens=x.split('-')
    if len(tokens)==2:
        return (float(tokens[0]) + float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2['total_sqft']=df2['total_sqft'].apply(convert)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df2.dropna(inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['price_per_sqft']=round(df2['price']*100000/df2['total_sqft'],2)
```

```python
df2
```

```python
df2.location=df2.location.apply(lambda x:x.strip())
loc_state=df2.groupby('location')['location'].count().sort_values(ascending=False)
loc_state
```

```python
loc_state[loc_state<=10]
```

```python
a=loc_state[loc_state<=10]
```

```python
df2['location']=df2['location'].apply(lambda x:'other' if x in a else x)
```

```python
df2['location'].nunique()
```

```python
df2.describe()
```

```python
# for 1 bhk we need 350 sqft
```

```python
df2[df2['total_sqft']/df2['BHK']<350]
```

```python
df3=df2[~(df2['total_sqft']/df2['BHK']<350)]
```

```python
df3
```

```python
df3['price_per_sqft'].describe()
```

```python
Q1 = df3['price_per_sqft'].quantile(0.25)
Q3 = df3['price_per_sqft'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound
```

```python
upper_bound
```

```python
df4 = df3[(df3['price_per_sqft'] >= lower_bound) & (df3['price_per_sqft'] <= upper_bound)]
```

```python
df4.shape
```

```python
df4
```

```python
cat_cols=['location']
df5=pd.get_dummies(df4,columns=cat_cols)
```

```python
df5
```

```python
X=df5.drop('price',axis=1)
y=df5['price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```
